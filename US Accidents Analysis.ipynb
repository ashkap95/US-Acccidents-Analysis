### Start Time

```python
df.Start_Time
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
sns.distplot(df.Start_Time.dt.hour, bins = 24, kde = False, norm_hist = True)
```

A high perecntage of accidents occur between 2 pm to 5 pm

```python
sns.distplot(df.Start_Time.dt.dayofweek, bins = 7, kde = False, norm_hist = True)
```

Is the distribution of accidents by hour same on the weekends as on weekdays?

```python
Sundays_start_time = df.Start_Time[df.Start_Time.dt.day_of_week == 6]
```

```python
sns.distplot(Sundays_start_time.dt.hour, bins = 24, kde = False, norm_hist = True)
```

```python
Monday_start_time = df.Start_Time[df.Start_Time.dt.day_of_week == 0]
sns.distplot(Monday_start_time.dt.hour, bins = 24, kde = False, norm_hist = True)
```

On weekdays there are more number of accidents during between 6 am to 8 am as compared to Sundays. On sundays there more accidents between 6 pm to 11 pm as compared to Mondays. The peak occurs between 1 pm to 6pm on sundays whereas it is between 3 pm to 5 pm on Mondays. 

```python
df_2018 = df[df.Start_Time.dt.year == 2018]
```

```python

sns.distplot(df_2018.Start_Time.dt.month, bins = 12, kde = False, norm_hist = True)
```
